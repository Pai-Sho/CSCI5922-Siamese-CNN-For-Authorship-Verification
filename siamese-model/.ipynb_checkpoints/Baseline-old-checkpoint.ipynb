 - Baseline architecture is designed with few layers and large kernels
    - 1st convolutional layer uses 32 filters of size 10 x 10 (32, 1, 10, 10)
    - 2nd convolutional layer uses 64 filters of size 8 x 8 (64, 1, 8, 8)
    - 3rd convolutional layer uses 64 filters of size 4 x 4 (64, 1, 4, 4)
    - 1st hidden layers uses 400 neurons (Relu)
    - 2nd hidden layer uses 200 hidden units (Relu)
    - L2 regularization and dropout with a probability of 0.5

```python
import torchvision
import torchvision.datasets as dset
import torch
import torch.nn as nn
from torch.utils.data import DataLoader,Dataset
from matplotlib.image import imread
import numpy as np
import torch.nn.functional as F
import torchvision.transforms as transforms
import pickle
from Dataset import AuthorsDataset
```

```python
class SiameseNet(nn.Module):
    def __init__(self):        
        super(SiameseNet, self).__init__()
        self.CNN = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=10),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, kernel_size=8), 
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),
            nn.Conv2d(64, 64, kernel_size=4), 
            nn.ReLU(inplace=True)
        )
        self.FC  = nn.Sequential(
            nn.Linear(64*253*1000, 400),
#             nn.Linear(64*2260*337, 400),
            nn.ReLU(inplace=True),
            nn.Linear(400, 200),
            nn.ReLU(inplace=True),
            nn.Linear(200, 2)
        )
        
    def forward_once(self, X):
        cnn_output = self.CNN(X)
        print(type(cnn_output))
        ANN_output = self.FC(cnn_output)
        return ANN_output
    
    def forward(self, image1, image2):
        image1_encoding = self.forward_once(image1)
        image2_encoding = self.forward_once(image2)
        
        return image1_encoding, image2_encoding
```
